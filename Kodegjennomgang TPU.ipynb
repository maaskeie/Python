<h2><font size="3" color="#000000">Avlesning av et exel-ark til GeoDataFrame</font>

<font size="2" color="#000000">  <i>OBS</i>  Viktig at navnene på arkfane, kolonner og lignende samsvarer <u>bokstavelig</u> med innholdet i det aktuelle exel-arket. Husk at Python er "case-sensitiv"/skiller mellom stor og liten bokstav.</font>

<h3><font size="2" color="#000000">Leser av Excel-dokumentet (.xlsx)</font>

```python
arkfane = 'Bergen_Floro'
geometrikolonne = 'punkt_geom_wkt'
```

<font size="2" color="#000000">Leser av exel-arket ved å bruke en standardfunksjon fra Pandasbiblioteket:</font>

```python
#Lager et leseobjekt for excelfiler
punkter = pd.read_excel(path, sheet_name=arkfane)
```

<font size="2" color="#000000">Rydder litt i mottatt DataFrame. Fjerner alle rader uten geometri og fjerner "whitespace"/"mellomrom" i kolonnenavnene</font>

```python
#dropper kolonner uten geometri (menon-logikk, ikke vurdert av oss)
punkter = punkter.dropna(subset=[geometrikolonne])

#I denne fila var det "whitespace" i kolonnenavnene, dette strippes bort her
punkter.columns = [col.strip() for col in list(punkter)]
```

<h3><font size="2" color="#000000">DataFrame til GeoDataFrame</font>

<font size="2" color="#000000">Nå eksisterer det en kolonne med geometrisk informasjon, men siden det er en DataFrame har den ikke verktøy for å forstå dette enda. Derfor må vi definere hvilken kolonne som inneholder geometri. <br><br>
Geometrien i dette dokumentet er representert ved en WKT-string ("Well Known Text"). Siden geometrikolonnen i GeoDataFrames ALLTID er shapelyfigurer benytter vi shapelybiblioteket til å konvertere geometrikolonnen.</font>

```python
#Får geomtrien gjort om til "Well Known Text"-representasjon bedre kjent som wkt.
punkter[geometrikolonne] = punkter[geometrikolonne].apply(wkt_loads)
```

<font size="2" color="#000000">Tar en titt på tabellen:</font>

```python
punkter.head(5)
```

<font size="2" color="#000000">Gjør DataFrame(Pandas) til GeoDataFrame(Geopandas) nå som vi har en geometrikolonne</font>

```python
#Redefinerer punkter fra DataFrame til GeoPandas. 
punkter = gpd.GeoDataFrame(punkter, geometry=geometrikolonne)
```

<font size="2" color="#000000">Dokumentet inneholder tiltakspunkter for hele Norge, så ønsker å begrense datasettet til ønsket region. </font>

```python
#Begrenser til kun punkter fra region Vest. String er bestemt av benyttede navn i .xlsx dokumentet
punkter=punkter[punkter.region == 'Vest']
```

<font size="2" color="#000000">Nå som vi har en geometrikolonne må vi definere gjeldende projeksjon. Vi har hentet et datasett som bruker EPSG 4326. Dette definerer datum, sone og format (wgs84, 33N, geografiske). </font>

```python
punkter.crs = {'init': 'epsg:4326'}
```

```python
punkter.head(50)
```

<h3><font size="2" color="#000000">bounding_box()</font>

<font size="2" color="#000000">Bounding box er begrep som beskriver de to koordinatene man trenger for å beskrive den minste rektangulære form som omslutter alle punktene. Disse uttrykkes som et sett av to punkter: (xmin, ymin, xmax, ymax)<br></font>
<font size="1" color="#000000">Kort forklart er (xmin, ymin) nedre høyre hjørne i en firkant, her kalt "bounding box". (xmax, ymin) blir nedre venstre hjørnet, (xmax, ymax) øverste til høyre og (xmin, ymax) øverst til venstre</font>

```python
boks = bounding_box(punkter)
```

<h3><font size="2" color="#000000">expand_box()</font>

<font size="2" color="#000000">Bounding box returnerer minste mulig rektangulære form, men fremdeles få med all geometri. Kan være litt liten i betydning at mange punkter kommer "nærme" kanten av utsnittet, lite kjekt ved visualisering og gir "edge-caser" ved analyse. Derfor kan man utvide den med ønsket prosentfaktor. Første argument er forrige boks og andre argument er ønsket utvidelse i prosentpoeng</font>

```python
utvidet_boks = expand_box(boks, 50)
```

<font size="2" color="#000000">Noen ganger vil man definere egen bounding boks, det kan gjøres så enkelt som følgende kodesnutt</font>
